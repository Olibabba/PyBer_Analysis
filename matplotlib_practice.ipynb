```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="d", color="purple", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.grid()
ax.set_ylim(0, 45)
ax.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label='Boston')
ax.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.barh(x_axis, y_axis, color="cyan", linewidth=2, label='Chicago')
ax.legend()
ax.invert_yaxis()
```

```python
# Create the scatter.
plt.plot(y_axis, x_axis, "o", c="red", label='Chicago')
# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Set the y limit between 0 and 45.
plt.xlim(0, 50)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis,  s = [i * 5 for i in y_axis], c="skyblue", alpha=.8, edgecolors='black', linewidth=2, label="Boston")
ax.legend()
ax.invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
```
